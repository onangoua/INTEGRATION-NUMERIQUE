```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
class Measure:
        # Constructeur de la classe
    def __init__(self, rayon):
        self.rayon = rayon
        self.a = 1
        self.b = 3
        self.c = 1
        self.d = 3
        self.x0 = 2
        self.y0 = 2
        
    def random(self, a, b, n):
        points = np.random.rand(n)
        points = (1 - points) * a + points * b
        return(points)
    
    def compute(self, n = 500):
        X = self.random(self.a, self.b, n)
        Y = self.random(self.c, self.d, n)
        n0 = 0
        areas = []
        plt.figure(figsize = (12, 6))
        plt.subplot(1, 2, 1)
        plt.xlim(0, 4)
        plt.ylim(0, 4)
        circle = plt.Circle((2, 2), 1, color = "darkgreen", linewidth = 1.5, fill = False)
        plt.gca().add_patch(circle)
        x_values = [1, 1, 3, 3, 1]
        y_values = [1, 3, 3, 1, 1]
        plt.plot(x_values, y_values, color = "darkred", linewidth = 1.5)
        for i in range(n):
            temp = (X[i] - self.x0)**2 + (Y[i] - self.y0)**2
            if (temp <= (self.rayon**2)):
                n0 = n0 + 1
                plt.scatter(X[i], Y[i], s = 2, c = "green", cmap = "summer")
            else:
                plt.scatter(X[i], Y[i], s = 2, c = "red", cmap = "summer")
            
            areas.append(np.round(n0/(i + 1), 5))                
        
        square_area = (self.b - self.a) * (self.d - self.c)
        areas = np.array(areas) * square_area
        plt.subplot(1, 2, 2)
        plt.xlim(0, n+1)
        plt.ylim(0, 5)
        pi = np.full(len(areas), np.pi)
        plt.plot(areas, c = "darkblue")
        plt.plot(pi, c = "darkred")
        plt.show()
        return(areas[-1])
```

```python
mes = Measure(1)
areas = mes.compute(101)
print(areas)
```
